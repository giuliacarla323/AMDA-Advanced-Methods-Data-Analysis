```python
# ==============================================================================
# EXPERIMENT 3: MULTI-LAYER PERCEPTRON (NEURAL NETWORK)
# ==============================================================================

# 1. INSTALARE SI IMPORTURI
!pip install imbalanced-learn -q

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, confusion_matrix, roc_curve)

# --- LINIA CARE LIPSEA ---
from sklearn.pipeline import Pipeline
# -------------------------

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# ==============================================================================
# 2. ÃŽNCÄ‚RCAREA DATELOR
# ==============================================================================

# âš ï¸ VerificÄƒ calea!
file_path = '/content/drive/MyDrive/AMDA/alzheimers_disease_data.csv'

try:
    df = pd.read_csv(file_path)
    print("âœ… Date Ã®ncÄƒrcate.")
except FileNotFoundError:
    print("âŒ Eroare: FiÈ™ierul nu a fost gÄƒsit.")

# ==============================================================================
# 3. PREPROCESARE (CRITIC PENTRU MLP!)
# ==============================================================================

# CurÄƒÈ›are de bazÄƒ
if 'PatientID' in df.columns:
    df = df.drop(columns=['PatientID'])

y = df['Diagnosis']
X = df.drop(columns=['Diagnosis'])
cols_to_drop = ['DoctorInCharge', 'Doctor', 'XXXConfid']
X = X.drop(columns=[c for c in cols_to_drop if c in X.columns])

# Identificare coloane (la fel ca Ã®nainte)
cat_cols_manual = ['Gender', 'Ethnicity', 'EducationLevel', 'Smoking',
                   'FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes',
                   'Depression', 'HeadInjury', 'Hypertension', 'MemoryComplaints',
                   'BehavioralProblems', 'Confusion', 'Disorientation',
                   'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness']

all_num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
all_cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

final_cat_cols = list(set(all_cat_cols + [c for c in cat_cols_manual if c in X.columns]))
final_num_cols = [c for c in all_num_cols if c not in final_cat_cols]

# Split Stratificat
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

# --- DEFINIREA TRANSFORMERELOR ---
# âš ï¸ PENTRU MLP, SCALAREA ESTE OBLIGATORIE (StandardScaler)
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())  # <--- ESENÈšIAL PENTRU NEURAL NETWORKS
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, final_num_cols),
    ('cat', categorical_transformer, final_cat_cols)
])

# ==============================================================================
# 4. CONFIGURARE MODEL MLP È˜I GRID SEARCH
# ==============================================================================

# Definim MLP de bazÄƒ
# max_iter=1000 asigurÄƒ cÄƒ reÈ›eaua are timp sÄƒ Ã®nveÈ›e (sÄƒ conveargÄƒ)
# early_stopping=True opreÈ™te antrenarea dacÄƒ nu se mai Ã®mbunÄƒtÄƒÈ›eÈ™te, prevenind overfitting
mlp = MLPClassifier(random_state=RANDOM_STATE, max_iter=1000, early_stopping=True)

pipeline_mlp = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=RANDOM_STATE)),
    ('clf', mlp)
])

# Parametrii pentru optimizare ("Tuning")
# Aici demonstrÄƒm cunoÈ™tinÈ›e avansate despre arhitecturi
param_grid = {
    # Arhitectura: (100,) = 1 strat ascuns cu 100 neuroni
    # (50, 25) = 2 straturi ascunse (Deep Learning), primul 50, al doilea 25
    'clf__hidden_layer_sizes': [(100,), (50, 25), (100, 50)],

    # FuncÈ›ia de activare: ReLU este standard, Tanh e clasic
    'clf__activation': ['relu', 'tanh'],

    # Regularizare L2 (Alpha) pentru a preveni overfitting-ul
    'clf__alpha': [0.0001, 0.01],

    # Rata de Ã®nvÄƒÈ›are: Cum face paÈ™ii gradient descent
    'clf__learning_rate_init': [0.001, 0.01]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

print("ðŸ§  ÃŽncepere antrenare ReÈ›ea NeuronalÄƒ (MLP)...")
print("Acest pas poate dura cÃ¢teva minute.")
grid_mlp = GridSearchCV(pipeline_mlp, param_grid, cv=cv, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_mlp.fit(X_train, y_train)

print(f"âœ… Best CV ROC-AUC: {grid_mlp.best_score_:.4f}")
print(f"Best Params: {grid_mlp.best_params_}")

best_model_mlp = grid_mlp.best_estimator_

# ==============================================================================
# 5. EVALUARE È˜I METRICI
# ==============================================================================

y_pred = best_model_mlp.predict(X_test)
y_proba = best_model_mlp.predict_proba(X_test)[:, 1]

print("\n=== REZULTATE MLP (Neural Network) ===")
print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred):.4f}")
print(f"F1 Score:  {f1_score(y_test, y_pred):.4f}")
print(f"ROC AUC:   {roc_auc_score(y_test, y_proba):.4f}")

# ==============================================================================
# 6. VIZUALIZÄ‚RI AVANSATE
# ==============================================================================

# A. Matricea de Confuzie
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - MLP (Neural Net)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# B. Curba ROC
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'MLP (AUC = {roc_auc_score(y_test, y_proba):.3f})', color='blue', lw=2)
plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.title('ROC Curve - MLP')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# C. CURBA DE ÃŽNVÄ‚ÈšARE (LOSS CURVE) - Specific pentru Neural Networks
# Trebuie sÄƒ extragem modelul MLP din pipeline-ul antrenat
mlp_trained = best_model_mlp.named_steps['clf']

plt.figure(figsize=(8, 6))
plt.plot(mlp_trained.loss_curve_, color='purple', lw=2)
plt.title('Learning Curve (Loss over Epochs)')
plt.xlabel('Epochs (Iterations)')
plt.ylabel('Loss (Eroare)')
plt.grid(alpha=0.3)
plt.show()

print("Interpretare Loss Curve: O curbÄƒ care scade lin È™i se stabilizeazÄƒ indicÄƒ o Ã®nvÄƒÈ›are sÄƒnÄƒtoasÄƒ.")

# ==============================================================================
# ANALIZA DETALIATÄ‚ A REZULTATELOR GRID SEARCH (MLP)
# ==============================================================================

results_mlp = pd.DataFrame(grid_mlp.cv_results_)

param_cols = [col for col in results_mlp.columns if col.startswith('param_')]
cols_to_keep = param_cols + ['mean_test_score', 'std_test_score', 'rank_test_score']
comparison_mlp = results_mlp[cols_to_keep].copy()
comparison_mlp.rename(columns={'mean_test_score': 'Mean ROC-AUC'}, inplace=True)
comparison_mlp = comparison_mlp.sort_values(by='Mean ROC-AUC', ascending=False)

print("\n--- Top 5 ConfiguraÈ›ii MLP ---")
display(comparison_mlp.head(5))

# VIZUALIZARE
plt.figure(figsize=(10, 8))

# Label scurt
def make_label_mlp(row):
    layers = row['param_clf__hidden_layer_sizes']
    act = row['param_clf__activation']
    lr = row['param_clf__learning_rate_init']
    return f"L={layers}, Act={act}, LR={lr}"

comparison_mlp['Label'] = results_mlp.apply(make_label_mlp, axis=1)

# Top 15
top_15_mlp = comparison_mlp.head(15)

sns.barplot(data=top_15_mlp, x='Mean ROC-AUC', y='Label', palette='Blues_r')
plt.axvline(grid_mlp.best_score_, color='red', linestyle='--', label=f'Best: {grid_mlp.best_score_:.4f}')
plt.title('Figure 4.13: Hyperparameter Stability Analysis (MLP)')
plt.xlabel('Mean ROC-AUC')
plt.tight_layout()
plt.savefig('mlp_grid_comparison.png', bbox_inches='tight')
plt.show()

```
